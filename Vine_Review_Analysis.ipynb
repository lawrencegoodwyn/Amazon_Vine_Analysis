```python
#Import dependencies

import pandas as pd
```

```python
# Read in CSV

vine_reviews_df = pd.read_csv("vine_table.csv")
vine_reviews_df
```

```python
filtered_df = vine_reviews_df[vine_reviews_df['total_votes'] >= 20]
filtered_df
```

```python
filtered_2_df = filtered_df[filtered_df['helpful_votes']/filtered_df['total_votes']*100 >= 50]

filtered_2_df


```

```python
vine_reviews_df= filtered_2_df[filtered_2_df['vine'] == 'Y']
vine_reviews_df
```

```python
n_vine_reviews_df = filtered_2_df[filtered_2_df['vine'] == 'N']
n_vine_reviews_df
```

```python
paid_five_star_df = vine_reviews_df.loc[vine_reviews_df['star_rating'] == 5]
unpaid_five_star_df = n_vine_reviews_df.loc[n_vine_reviews_df['star_rating'] == 5]
```

```python
# Total Vine (paid) reviews, total 5-star reviews, % of 5 stars vs vine reviews

print(f'The total number of paid reviews is {len(vine_reviews_df)}')
print(f'The total number of 5-star reviews is {len(paid_five_star_df)}')
print(f'The percentage of paid 5 star reviews to total reviews is {len(paid_five_star_df) / len(vine_reviews_df) * 100}')
```

```python
# Total Vine (un-paid) reviews, total 5-star reviews, % of 5 stars vs regular reviews
print(f'The total number of paid reviews is {len(n_vine_reviews_df)}')
print(f'The total number of 5-star reviews is {len(unpaid_five_star_df)}')
print(f'The percentage of paid 5 star reviews to total reviews is {len(unpaid_five_star_df) / len(n_vine_reviews_df) * 100 %}')
```
